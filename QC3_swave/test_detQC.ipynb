```python
import plotly.express as px
from F3 import F3_mat as F3mat

NL=800
L=5.457
nnP=np.array([0, 0, 0])
params_kcot=np.array([-0.14169772205375078, -2.911948317634196])
ei=np.zeros(NL)
Fiso=np.zeros(NL)
Kiso=np.zeros(NL)
Gplot=np.zeros(NL)
def kcot(qk2): return detQC.kcot_2par(qk2, params_kcot)



for i in range(0,NL):
    ei[i]=1.01+i*0.003
    F00=F3mat.F3mat00(ei[i], L, 0.5, nnP, kcot)
    # Gplot[i]=
    ones = np.ones(len(F00))
    Fiso[i] = 1/(ones@F00@ones)
    ecm2 = (ei[i]**2 - (sums.norm(nnP)*2*math.pi/L)**2)
    Delta = (ecm2 - 9.0)/9.0
    #print(Delta,ei[i])
    Kiso[i] =  detQC.kiso_pole(Delta, [-83.38, 9.65658, -1664.41])
```

```python

fig = px.line(  title='QC3-pole')
 
fig.add_scatter(x=ei, y=Fiso, mode = 'lines+markers',name='Fiso{0}'.format(1)  )
fig.add_scatter(x=ei, y=-Kiso, mode = 'lines+markers',name='Kiso{0}'.format(1)  )
fig.add_scatter(x=ei, y=Fiso+Kiso, mode = 'lines+markers',name='QC3{0}'.format(1)  )
#plt.legend()
#fig.update_layout(yaxis_range=[-2000,3000])
#fig.update_layout(xaxis_range=[3,3.2])

fig.show()   
Fiso
```
